```python
x, y = 0, 0
for m, step in moves:
    if m == "forward":
        x += step
    elif m == "down":
        y += step
    elif m == "up":
        y -= step
    else:
        raise ValueError(f"unk: {m}")

mul = x * y
```

```python
x, y, aim = 0, 0, 0
for m, step in moves:
    if m == "forward":
        x += step
        y += aim * step
    elif m == "down":
        aim += step
    elif m == "up":
        aim -= step
    else:
        raise ValueError(f"unk: {m}")
    print(m, step, x, y, aim)

mul = x * y
```
